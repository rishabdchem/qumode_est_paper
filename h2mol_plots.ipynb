```python
# Figure and axes
fig, ax = plt.subplots()
plt.rcParams.update({'font.size': 14})
plt.subplots_adjust(left=0.15, bottom=0.15)

# Axis labels and title
ax.set_xlabel('Bond distance ($\AA$)')
ax.set_ylabel('Energy (Hartree)')

# Plot the curves
ax.plot(dis_angstrom_fci, fci_en, '', color='black', label='FCI')
ax.plot(dis_angstrom_vqe, en_vqe_lcu_ecd_nd9, 'o', color='blue', label='ECD-VQE')

# Legend
ax.legend(loc='upper right', fontsize="15")

# Save
plt.savefig('vqe_en_ecd.pdf', dpi=500)

# Show the plot
plt.show()
```
